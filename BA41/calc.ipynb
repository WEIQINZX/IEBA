```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot |x-y| in 3D
def plot_3d(x, y):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, np.abs(x-y))
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('|x-y|')
    plt.show()

# plot |x-y| in 2D
def plot_2d(x, y):
    plt.plot(x, np.abs(x-y))
    plt.xlabel('x')
    plt.ylabel('|x-y|')
    plt.show()

# plot
def plot(x, y):
    plot_3d(x, y)
    plot_2d(x, y)

# draw 3d for x in the range of [0, 1]
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
plot(x, y)
```
